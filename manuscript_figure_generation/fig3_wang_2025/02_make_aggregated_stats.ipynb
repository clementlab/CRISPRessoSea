```python
with open("aggregated_stats_all.txt", 'r') as fin, open('02_make_aggregated_stats.ipynb.txt','w') as fout:
    head = fin.readline()
    fout.write(head)
    for line in fin:
        line_els = line.split("\t")
        name = '_'.join(line_els[1].split("_")[1:])
        line_els[1] = name
        fout.write('\t'.join(line_els))
```
